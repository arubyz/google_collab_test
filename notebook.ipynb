````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.sin(x**2)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='sin(x^2)')
plt.title('Plot of sin(x) and sin(x^2)')
plt.xlabel('x')
plt.ylabel('sin(x) and sin(x^2)')
plt.legend()
plt.grid(True)
plt.show()

# Successful test case
assert 2 == 2
````
